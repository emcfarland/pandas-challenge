```python
import pandas as pd

student_data_path = 'Resources/students_complete.csv'
school_data_path = 'Resources/schools_complete.csv'

student_data = pd.read_csv(student_data_path)
school_data = pd.read_csv(school_data_path)

complete_data = pd.merge(student_data, school_data, how='left', on=['school_name', 'school_name'])
```

```python
complete_data.head()
```

```python
school_names = complete_data['school_name'].unique()
tot_students = complete_data['student_name'].count()
tot_avg_math = complete_data['math_score'].mean()
tot_avg_reading = complete_data['reading_score'].mean()
pass_math = complete_data['math_score'].loc[complete_data['math_score']>=70]
pass_reading = complete_data['reading_score'].loc[complete_data['reading_score']>=70]
pass_overall = complete_data.loc[(complete_data['math_score']>=70) & (complete_data['reading_score']>=70)]


by_district_df = pd.DataFrame({
    'Total Schools': [school_names],
    'Total Students': tot_students,
    'Total Budget': sum(complete_data['budget'].unique()),
    'Average Math Score': tot_avg_math,
    'Average Reading Score': tot_avg_reading,
    '% Passing Math': 100*len(pass_math)/tot_students,
    '% Passing Reading': 100*len(pass_reading)/tot_students,
    '% Overall Passing': 100*len(pass_overall)/tot_students
})
by_district_df
```
